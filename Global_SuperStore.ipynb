# KPI SUMMARY

Key Performance Indicators (KPIs) provide a quick overview of the company's overall business performance. Metrics such as total sales, total profit, total orders, and average order value help establish the foundation for further analysis.

```python
total_sales=df['Sales'].sum()
total_profit=df['Profit'].sum()
total_orders=df['Order ID'].nunique()
total_customers=df['Customer ID'].nunique()
profit_margin=(total_profit/total_sales)*100
avg_order=total_sales/total_orders
```

```python
print("TOTAL SALES : ",total_sales)
print("TOTAL PROFIT : ",total_profit)
print("TOTAL ORDERS : ",total_orders)
print("TOTAL CUSTOMERS : ",total_customers)
print("PROFIT MARGIN : ",profit_margin)
print("AVERAGE ORDER VALUE : ",avg_order)
```

# Sales Trend Analysis 

This section analyzes how sales change over time to identify seasonal trends, growth patterns, and high-performing periods. Understanding these trends helps businesses improve forecasting and strategic planning.

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['year']=df['Order Date'].dt.year     #created column YEAR
```

```python
df['month']=df['Order Date'].dt.month_name
```

```python
df['year_month']=(df['Order Date'].dt.to_period('M'))
```

```python
monthly_sales=df.groupby('year_month')['Sales'].sum().reset_index() 
monthly_sales.head()
monthly_sales['year_month']=monthly_sales['year_month'].astype(str)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns 
plt.figure(figsize=(15, 6))
sns.lineplot(x='year_month', y='Sales', data=monthly_sales)
plt.title('Total Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales Value')
plt.xticks(rotation=45)  
plt.tight_layout()  
plt.show()
```

```python
df['quarter']=df['Order Date'].dt.quarter
```

```python
# how many transactions in each quarter 
df.groupby('quarter')['Order ID'].count()
```

## Business Recommendations

Based on the analysis performed in this project, the following recommendations can help improve business performance:

- **Review discount strategies:** Excessive discounts should be carefully monitored, particularly for low-margin products, to prevent unnecessary reductions in profitability.

- **Focus on high-performing categories:** Continue investing in product categories that consistently generate strong sales and profits while evaluating underperforming categories for improvement or restructuring.

- **Strengthen regional strategies:** Analyze the practices of high-performing regions and apply successful strategies to regions with lower sales or profitability.

- **Prioritize high-value customers:** Identify loyal and high-spending customers and develop targeted retention programs, personalized offers, and loyalty initiatives to maximize customer lifetime value.

- **Improve inventory planning:** Use historical sales trends to forecast demand more accurately, especially during peak sales periods, reducing stock shortages and excess inventory.

- **Monitor operational efficiency:** Evaluate shipping methods and associated costs regularly to balance customer satisfaction with operational profitability.

Implementing these recommendations can help improve revenue growth, operational efficiency, and long-term profitability while supporting data-driven business decision-making.
